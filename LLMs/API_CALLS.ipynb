## OpenAI sdk

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv() 

client = OpenAI()    # here client will be our object

response = client.responses.create(
    model="gpt-5-mini",
    input="Tell me fun fact about data domain"     #here response will be our object
)

print(response.output_text)
```
